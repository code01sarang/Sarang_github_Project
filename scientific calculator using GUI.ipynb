```python
import tkinter as tk
import math
```

```python
window = tk.Tk()
window.title("Simple Calculator")
```

```python

def addition():
    a=float(entry1.get())
    b=float(entry2.get())
    c=a+b
    entry3.insert(0,c)
    
def substraction():
    a=float(entry1.get())
    b=float(entry2.get())
    c=a-b
    entry3.insert(0,c)

def multiplication ():
    a=float(entry1.get())
    b=float(entry2.get())
    c=a*b
    entry3.insert(0,c)

def division():
    a=float(entry1.get())
    b=float(entry2.get())
    c=a/b
    entry3.insert(0,c)

def modulus():
    a=float(entry1.get())
    b=float(entry2.get())
    c=a%b
    entry3.insert(0,c)

def sin_func():
    a = float(entry1.get())
    c = math.sin(math.radians(a))
    entry3.insert(0,c)

def cos_func():
    a = float(entry1.get())
    c = math.cos(math.radians(a))
    entry3.insert(0,c)

def tan_func():
    a = float(entry1.get())
    c = math.tan(math.radians(a))
    entry3.insert(0,c)
```

```python
def delete():
    entry1.delete(0,END)
    entry2.delete(0,END)
    entry3.delete(0,END)
    
```

```python
label1=Label(window,text="Enter No.1: ")
label2=Label(window,text="Enter No.2: ")
```

```python
#window.mainloop()
```

```python
tk.Label(window, text="Enter No.1: ").grid(row=0, column=0)
entry1 = tk.Entry(window)
entry1.grid(row=0, column=1)

tk.Label(window, text="Enter No.2: ").grid(row=1, column=0)
entry2 = tk.Entry(window)
entry2.grid(row=1, column=1)

tk.Label(window, text="Result").grid(row=2, column=0)
result = tk.Entry(window)
result.grid(row=2, column=1)
```

```python
entry1=Entry(window)
entry2=Entry(window)
add=Button(window,text="ADD",command=addition)
sub=Button(window,text="SUB",command=substraction)
mul=Button(window,text="MUL",command=multiplication)
div=Button(window,text="DIV",command=division)
mod=Button(window,text="MOD",command=modulus)
sin=Button(window, text="Sin", command=sin_func)
cos=Button(window, text="Cos", command=cos_func)
tan=Button(window, text="tan", command=tan_func)


entry3=Entry(window)
clear=Button(window,text="CLEAR",command=delete)
```

```python
label1.grid(row=0,column=0)
label2.grid(row=1,column=0)
entry1.grid(row=0,column=1)
entry2.grid(row=1,column=1)
add.grid(row=2,column=0)
sub.grid(row=3,column=0)
mul.grid(row=4,column=0)
div.grid(row=5,column=0)
mod.grid(row=6,column=0)
sin.grid(row=7,column=0)
cos.grid(row=8,column=0)
tan.grid(row=9,column=0)

entry3.grid(row=2,column=1)
clear.grid(row=10,column=0)
```

```python
window.mainloop()
```
